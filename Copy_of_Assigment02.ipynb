```python
#Ques 2
#Importing seaborn as sns
import seaborn as sns
#Importing plot package
import matplotlib.pyplot as plt
#Importing Statistics package
import statistics
#Storing all acceptable Noise limit data in variable Accp_NL
Accp_NL=[55.3,55.3,55.3,55.9,55.9,55.9,55.9,56.1,56.1,56.1,56.1,56.1,56.1,56.8,56.8,57.0,57.0,57.0,57.8,57.8,
57.8,57.9,57.9,57.9,58.8,58.8,58.8,59.8,59.8,59.8,62.2,62.2,63.8,63.8,63.8,63.9,63.9,63.9,64.7,64.7,
64.7,65.1,65.1,65.1,65.3,65.3,65.3,65.3,67.4,67.4,67.4,67.4,68.7,68.7,68.7,68.7,69.0,70.4,70.4,71.2,
71.2,71.2,73.0,73.0,73.1,73.1,74.6,74.6,74.6,74.6,79.3,79.3,79.3,79.3,83.0,83.0,83.0]
#Total sample
n=len(Accp_NL)
print(n)
#(a)Summarizing the Data
#Measure of Central Tendency
#Mean
#Initialising sum as 0
sum=0
#Using for loop to sum each elemet of list
for i in Accp_NL:
  sum=sum+i
mean=sum/n
print("Mean is:",mean)
#Median
median=statistics.median(Accp_NL)
print("Median is:",median)
#Mode
mode=statistics.mode(Accp_NL)
print("Mode is:",mode)
#Measure of Dispersion
#1)Range
range=Accp_NL[n-1]-Accp_NL[0]
print("Range is:",range)
#Variance
var=statistics.variance(Accp_NL)
print("Variance is:",var)
#Standar Deviation
std=statistics.stdev(Accp_NL)
print("Standar Deviation is:",std)
#Mean Absolute Deviation
l= []
for x in Accp_NL:
  dif=abs(x-mean)
  l.append(dif)
MAD=statistics.mean(l)
print("Mean Absolute Deviation is:",MAD)
#Median Absolute Deviation
Me_AD=statistics.median(l)
print("Median Absolute Deviation is:",Me_AD)
#Distribution of Data
print("Data Distribution")
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.hist(Accp_NL)
plt.xlabel('Noise Level (dBA)')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
#Box Plot
plt.figure(figsize=(15,5))
plt.subplot(132)
plt.boxplot(Accp_NL)
plt.xlabel('Noise Level (dBA)')
plt.ylabel('Frequency')
plt.title('Box Plot')
plt.show()
#Violin Plot
plt.figure(figsize=(15,5))
plt.subplot(133)
sns.violinplot(Accp_NL,inner='box')
plt.ylabel('Noise Level (dBA)')
plt.ylabel('Frequency')
plt.title('Violin Plot')
plt.show()
plt.tight_layout()
plt.show()
















```
